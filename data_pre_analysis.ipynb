### 3.2 Analysis: calculating the difference between fragments

Creating the graph used for comparing false positives and false negatives.

```python
def get_count_percentage (doc):
    total = 0
    count_hpo3 = 0
    count_h3po4 = 0
    count_ser = 0
    count_tyr = 0
    count_thr = 0
    
    for ms1 in doc:
        total += 1
        if ms1.HPO3[0] is not None:
            count_hpo3 += 1 
       
        if ms1.H3PO4[0] is not None:
            count_h3po4 += 1 
        if ms1.ser[0] is not None:
            count_ser += 1 
        if ms1.tyr[0] is not None:
            count_tyr += 1 
        if ms1.thr[0] is not None:
            count_thr += 1 
            
    print count_hpo3, count_h3po4
    count_hpo3_percentage = percent(count_hpo3, total)
    count_h3po4_percentage = percent(count_h3po4, total)
    count_ser_percentage = percent(count_ser, total)
    count_tyr_percentage = percent(count_tyr, total)
    count_thr_percentage = percent(count_thr, total)
    
    return ((count_hpo3_percentage,count_h3po4_percentage, count_ser_percentage,
            count_tyr_percentage, count_thr_percentage))
```

```python
ppm_list = [0, 2, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 100, 150, 200, 300, 400]

phos_count_list = []
nonphos_count_list = []

for ppm in ppm_list:
    analyser_phos.add_theoretical_seq_ion(seq_tol=ppm/2, diff_tol=ppm)
    analyser_nonphos.add_theoretical_seq_ion(seq_tol=ppm/2, diff_tol=ppm) 

    document_list1 = analyser_phos.document_list
    document_list2 = analyser_nonphos.document_list

    phos_count_list.append(get_count_percentage(document_list1))                     
    nonphos_count_list.append(get_count_percentage(document_list2))
```

```python
trace0 = go.Scatter(x = ppm_list,y = [hpo3_count for hpo3_count,_,_,_,_ in phos_count_list],
    mode = 'lines+markers',name = 'HPO3',
    marker=dict(color='#FF6347',line=dict(width=1),symbol='square',size=6))

trace1 = go.Scatter(x = ppm_list,y = [hpo3_count for hpo3_count,_,_,_,_ in nonphos_count_list],
    mode = 'lines+markers',name = 'HPO3 (non-PTM dataset)',
    marker=dict(color='#FF6347',line=dict(width=1),symbol='circle',size=6))

trace2 = go.Scatter(x = ppm_list,y = [h3po4_count for _,h3po4_count,_,_,_ in phos_count_list],
    mode = 'lines+markers',name = 'H3PO4',
    marker=dict(color='BA55D3',line=dict(width=1),symbol='square',size=6,))

trace3 = go.Scatter(x = ppm_list,y = [h3po4_count for _,h3po4_count,_,_,_ in nonphos_count_list],
    mode = 'lines+markers',name = 'H3PO4 (non-PTM dataset)',
    marker=dict(color='BA55D3',line=dict(width=1),symbol='circle',size=6))

trace4 = go.Scatter(x = ppm_list,y = [ser_count for _,_,ser_count,_,_ in phos_count_list],
    mode = 'lines+markers',name = 'ser',
    marker=dict(color='FFB6C1',line=dict(width=1),symbol='square',size=6))

trace5 = go.Scatter(x = ppm_list,y = [ser_count for _,_,ser_count,_,_ in nonphos_count_list],
    mode = 'lines+markers',name = 'ser (non-PTM dataset)',
    marker=dict(color='FFB6C1',line=dict(width=1),symbol='circle',size=6))

trace6 = go.Scatter(x = ppm_list,y = [tyr_count for _,_,_,tyr_count,_ in phos_count_list],
    mode = 'lines+markers',name = 'tyr',
    marker=dict(color='E6E6FA',line=dict(width=1),symbol='square',size=6))

trace7 = go.Scatter(x = ppm_list,y = [tyr_count for _,_,_,tyr_count,_ in nonphos_count_list],
    mode = 'lines+markers',name = 'tyr (non-PTM dataset)',
    marker=dict(color='E6E6FA',line=dict(width=1),symbol='circle',size=6))

trace8 = go.Scatter(x = ppm_list,y = [thr_count for _,_,_,_,thr_count in phos_count_list],
    mode = 'lines+markers',name = 'thr',
    marker=dict(color='F5F5DC',line=dict(width=1),symbol='square',size=6))

trace9 = go.Scatter(x = ppm_list,y = [thr_count for _,_,_,_,thr_count in nonphos_count_list],
    mode = 'lines+markers',name = 'thr (non-PTM dataset)',
    marker=dict(color='F5F5DC',line=dict(width=1),symbol='circle',size=6))

data = [trace0,trace2,trace4,trace6,trace8,trace1,trace3,trace5,trace7,trace9]

layout = go.Layout(xaxis=dict(title='Tolerance (ppm)'),
                   yaxis=dict(range = [0,100],
                              title='MS1 spectra with specific loss (%)'))

fig = go.Figure(data=data, layout=layout)
plot_url = py.plot(fig, filename='test_tolerance')

py.iplot(fig, filename='scatter-mode')

```
